## **Ejercicio N°1:**
Escribir un programa que pregunte al usuario por las ventas de los últimos 5 días, y muestre por pantalla un Dataframe con los datos de las ventas por producto: Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.


```python
import pandas as pd
import random

print("############ Por favor, ingresar las ventas de los ultimos 5 días #############")

ventas=[]
diccionario={"Hace_x_dia":[],"codigo_producto":[],"unidades":[], "precio_unitario":[]}

for i in range(1,6):
  ventas.append(input("Ingrese la cantidad de ventas de hace " + str(i)+ " día(s): "))

m=0
for cantidad_ventas in ventas:
  m+=1
  for x_producto in range(int(cantidad_ventas)):
    diccionario["Hace_x_dia"].append(m)
    diccionario["codigo_producto"].append(random.randint(100000,999999))
    diccionario["unidades"].append(random.randint(1,100))
    diccionario["precio_unitario"].append(random.randint(100,999))

df=pd.DataFrame(data=diccionario)
df=df.set_index("Hace_x_dia")
df["Importe Total"]=df["unidades"]*df["precio_unitario"]
df["Importe Total con descuento"]=df["Importe Total"]*0.5
df
```

## **Ejercicio 2**
Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.


```python
import pandas as pd 

def insights(ingresos):
  ingresos_serie=pd.Series(ingresos)
  estadisticos=pd.Series([ingresos_serie.min(),ingresos_serie.max(), ingresos_serie.mean()], index=["Ingreso_Min", "Ingreso_max", "Ingreso_medio"])
  return estadisticos

ingresos={"Felipe":4000, "Yomira":5000, "Edgar":500, "Paola":6000, "Luis":700}

print(insights(ingresos))
```

# **Ejercicio 3**
Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:


Mes	Ventas	Gastos

Enero	30500	22000

Febrero	35600	23450

Marzo	28300	18100

Abril	33900	35700

Mayo	42500	32450

A continuación, generar nuevas columna:
1.	Columna 1 =  Gasto/Ventas*100.
2.	Columna 2 =  Si Ventas>= 3000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3.	Columna 3 =  Si Gasto>Ventas, calcular el sobre gasto.
4.	Obtener el total de las ventas de los últimos 3 meses
5.	Obtener el total de gastos de febrero a abril


```python
import pandas as pd
DataFrame=pd.DataFrame()

DataFrame["Mes"]=["Enero", "Febrero", "Marzo", "Abril", "Mayo"]
DataFrame["Ventas"]=[30500,35600,28300,33900,42500]
DataFrame["Gastos"]=[22000,23450,18100,35700,32450]

DataFrame

```

```python
#Columna 1 = Gasto/Ventas*100.
DataFrame["columna1"]=DataFrame["Gastos"]/DataFrame["Ventas"]*100
DataFrame
```

```python
#Columna 2 = Si Ventas>= 3000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
DataFrame["Meta"]=["Meta superada" if venta>=30000 else "Meta no superada" for venta in DataFrame["Ventas"]]
DataFrame
```

```python
#Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
import numpy as np
x=DataFrame["Gastos"]-DataFrame["Ventas"]
DataFrame["Sobre gasto"]=np.where(x>0, x, "No hay sobre gasto")
DataFrame
```

```python
#Obtener el total de las ventas de los últimos 3 meses
DataFrame.loc[2:,"Ventas"].sum()
```

```python
#Obtener el total de gastos de febrero a abril
DataFrame.loc[1:3,"Gastos"].sum()
```
